```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
def graph_plot(df,flag1,flag2,title):
    clf = ['a','aa','rr','a','aa','rr','a','aa']
    labels = df.iloc[:,0]
    if flag1==1:
        data1 = df.iloc[:,1]
        data2 = df.iloc[:,3]
        data3 = df.iloc[:,5]
        data4 = df.iloc[:,7]
        data5 = df.iloc[:,9]
        data6 = df.iloc[:,11]
        data7 = df.iloc[:,13]
        data8 = df.iloc[:,15]
    elif flag1==2:
        data1 = df.iloc[:,2]
        data2 = df.iloc[:,4]
        data3 = df.iloc[:,6]
        data4 = df.iloc[:,8]
        data5 = df.iloc[:,10]
        data6 = df.iloc[:,12]
        data7 = df.iloc[:,14]
        data8 = df.iloc[:,16]
    elif flag1==3:
        data1 = df.iloc[:,1]
        data2 = df.iloc[:,3]
        data3 = df.iloc[:,5]
        data4 = df.iloc[:,7]
        data5 = df.iloc[:,9]
        data6 = df.iloc[:,11]
        data7 = df.iloc[:,13]
    elif flag1==4:
        data1 = df.iloc[:,2]
        data2 = df.iloc[:,4]
        data3 = df.iloc[:,6]
        data4 = df.iloc[:,8]
        data5 = df.iloc[:,10]
        data6 = df.iloc[:,12]
        data7 = df.iloc[:,14]
        
        
    if flag1==1:
        groups = [data1, data2,data3,data4, data5,data6,data7,data8] 
    elif flag1==2:
        groups = [data1, data2,data3,data4, data5,data6,data7,data8]
    elif flag1==3:
        groups = [data1, data2,data3,data4, data5,data6,data7] 
    elif flag1==4:
        groups = [data1, data2,data3,data4, data5,data6,data7] 
      
    margin = 0.009
    width = (1.-2.*margin)/10
    ind = np.arange(10)
    s = plt.subplot(1,1,1)
    plt.legend(labels)
    for num, vals in enumerate(groups):
        # The position of the xdata must be calculated for each of the two data series
        xdata = ind+margin+(num*width)
        # Removing the "align=center" feature will left align graphs, which is what
        # this method of calculating positions assumes
        gene_rects = plt.bar(xdata, vals, width)

        
        
    if (flag1==1 or flag1==2) and flag2==1:
        plt.legend(['KNN','SVM','DTC','RFC','ABC','LR','GNB','NN'])
        
    elif (flag1==3 or flag1==4) and flag2==1:
        plt.legend(['KNN','SVM','DTC','RFC','ABC','LR','NN'])
      
    elif (flag1==1 or flag1==2) and flag2==2:
        plt.legend(['SVR_RBF' ,'SVRLinear','DTR','RFR','ABR','GPR','LIR','RIR'])    
        
    elif (flag1==3 or flag1==4) and flag2==2:
         plt.legend(['SVRSVR_RBF' ,'SVRLinear','DTR','RFR','ABR','GPR','RIR'])
        
    
   
    
    
    # You should no longer need to manually set the plot limit since everything 
    # is scaled to one.
    # Also the ticks should be much simpler now that each group of bars extends from
    # 0.0 to 1.0, 1.0 to 2.0, and so forth and, thus, are centered at 0.5, 1.5, etc.
    s.set_xticks(ind+0.5)
    s.set_xticklabels(labels,rotation=90)
    s.set_title(title)
```

```python
df = pd.read_csv('ML_Project - Plot_Cls_default.csv',skiprows=1)
graph_plot(df,1,1,"Classification Training Accuracy (default)")
```

```python
graph_plot(df,2,1,"Classification Testing Accuracy (default)")
```
